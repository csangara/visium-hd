```python
%matplotlib inline
from matplotlib import pyplot as plt

import numpy
import json

with open('/my/Surfdrive/Shared/EP23/EP23_clahe_qn_va/visualign_export/EP23_orig42_s031_nl.flat','rb') as fp:
  buffer = fp.read()
nDims = int(buffer[0])
shape = numpy.frombuffer(buffer, dtype=numpy.dtype('>i4'), offset=1, count=2) 
data = numpy.frombuffer(buffer, dtype=numpy.dtype('>i2'), offset=9)
data = data.reshape(shape[::-1])
print(nDims,shape,data.shape,data.min(),data.max())

plt.imshow(data)
```

```python
with open('/my/Surfdrive/Shared/EP23/EP23_clahe_qn_va/slice_exported/Rainbow 2017.json') as fp:
  colors = json.load(fp)
  
R = numpy.zeros(data.shape,numpy.uint8)
G = numpy.zeros(data.shape,numpy.uint8)
B = numpy.zeros(data.shape,numpy.uint8)
for c in numpy.unique(data):
  clr = colors[c]
  idx = (data==c)
  R[idx] = clr['red']
  G[idx] = clr['green']
  B[idx] = clr['blue']
rgb = numpy.stack((R,G,B),axis=2)
```
