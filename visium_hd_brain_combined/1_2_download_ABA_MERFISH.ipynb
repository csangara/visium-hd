```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
# Read in cell metadata with CCF coords
base_path = "/home/chananchidas/visium-hd/data/ABA_metadata"
```

```python
cell = pd.read_csv(base_path + "/MERFISH-C57BL6J-638850/20241115/views/cell_metadata_with_cluster_annotation.csv")
cell.rename(columns={'x': 'x_section',
                     'y': 'y_section',
                     'z': 'z_section'},
            inplace=True)
cell.set_index('cell_label', inplace=True)
cell.head(5)
```

```python
reconstructed_coords = pd.read_csv(base_path + "/MERFISH-C57BL6J-638850-CCF/20231215/reconstructed_coordinates.csv")
reconstructed_coords.rename(columns={'x': 'x_reconstructed',
                                     'y': 'y_reconstructed',
                                     'z': 'z_reconstructed'},
                            inplace=True)
reconstructed_coords.set_index('cell_label', inplace=True)
reconstructed_coords.head(5)
```

```python
ccf_coords = pd.read_csv(base_path + "/MERFISH-C57BL6J-638850-CCF/20231215/ccf_coordinates.csv")
ccf_coords.rename(columns={'x': 'x_ccf',
                           'y': 'y_ccf',
                           'z': 'z_ccf'},
                  inplace=True)
ccf_coords.drop(['parcellation_index'], axis=1, inplace=True)
ccf_coords.set_index('cell_label', inplace=True)
ccf_coords.head(5)
```

```python
cell_joined = cell.join(reconstructed_coords, how='inner')
cell_joined = cell_joined.join(ccf_coords, how='inner')
cell_joined.head(5)
```

```python
# Read in parcellation annotation
parcellation_annotation = pd.read_csv(base_path + "/Allen-CCF-2020/20230630/views/parcellation_to_parcellation_term_membership_acronym.csv")
parcellation_annotation.set_index('parcellation_index', inplace=True)
parcellation_annotation.columns = ['parcellation_%s'% x for x in  parcellation_annotation.columns]
parcellation_annotation.head(5)

parcellation_color = pd.read_csv(base_path + "/Allen-CCF-2020/20230630/views/parcellation_to_parcellation_term_membership_color.csv")
parcellation_color.set_index('parcellation_index', inplace=True)
parcellation_color.columns = ['parcellation_%s'% x for x in  parcellation_color.columns]
parcellation_color.head(5)
```

```python
cell_joined = cell_joined.join(parcellation_annotation, on='parcellation_index')
cell_joined = cell_joined.join(parcellation_color, on='parcellation_index')
cell_joined.head(5)
```
